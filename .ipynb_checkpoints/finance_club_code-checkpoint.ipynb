```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set up nicer plots
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['figure.dpi'] = 100
sns.set_style('whitegrid')

# Load the data
df = pd.read_csv('/Users/shabadpreetsingh/Downloads/Finance_Project/dataset.csv')
print(f'Shape: {df.shape[0]} rows x {df.shape[1]} columns')
df.head()
```

### Understanding the columns

- `datetime`: timestamp every 5 minutes during market hours (9:15–15:30) from 7 Jan to 27 Jan 2026.
- `underlying_price`: the spot price of Nifty50 at that moment.
- The other 28 columns are option IVs. Each name follows the pattern `NIFTY27JAN26<strike><CE|PE>`.
  - `NIFTY` = the index
  - `27JAN26` = expiry date (27 January 2026 — same for all columns)
  - `<strike>` = the strike price (e.g. 25900)
  - `CE` or `PE` = Call or Put

```python
# Parse datetime and extract strike info
df['datetime'] = pd.to_datetime(df['datetime'], format='%d-%m-%Y %H:%M')

ce_cols = sorted([c for c in df.columns if c.endswith('CE')])
pe_cols = sorted([c for c in df.columns if c.endswith('PE')])
opt_cols = pe_cols + ce_cols   # Puts first (lower strikes), then calls (higher strikes)

def get_strike(col):
    return int(col.replace('NIFTY27JAN26', '').replace('CE', '').replace('PE', ''))

strikes = [get_strike(c) for c in opt_cols]

print(f'Number of Call (CE) columns: {len(ce_cols)}')
print(f'Number of Put  (PE) columns: {len(pe_cols)}')
print(f'\nCall strikes range: {min([get_strike(c) for c in ce_cols])} to {max([get_strike(c) for c in ce_cols])}')
print(f'Put  strikes range: {min([get_strike(c) for c in pe_cols])} to {max([get_strike(c) for c in pe_cols])}')
print(f'\nUnderlying price range: {df["underlying_price"].min():.0f} to {df["underlying_price"].max():.0f}')
print(f'Underlying mean: {df["underlying_price"].mean():.0f}')
```

**Important observation:** Notice that the *put* strikes are all *below* 25200 and the *call* strikes are all *above* 25100. They don't overlap. The underlying price moves around 25000–26200.

This is normal — in real option chains, only the *out-of-the-money* (OTM) options are traded actively, so you only see puts below spot and calls above spot. The 'gap' near the spot price will be visible when we plot the smile.

## 3. Where are the missing values?

We need to predict missing values, so first understand WHERE they are and HOW they're distributed.

```python
missing_total = df[opt_cols].isnull().sum().sum()
total_cells = df[opt_cols].size

print(f'Total option cells: {total_cells}')
print(f'Missing cells:      {missing_total}')
print(f'Missing percentage: {missing_total/total_cells*100:.2f}%')
print()
print(f'Average missing per column: {df[opt_cols].isnull().sum().mean():.1f} (out of {df.shape[0]} rows)')
print(f'Average missing per row:    {df[opt_cols].isnull().sum(axis=1).mean():.1f} (out of {len(opt_cols)} columns)')
```

```python
# Visualize the missingness pattern
fig, ax = plt.subplots(figsize=(13, 6))
missing_mask = df[opt_cols].isnull().astype(int)
ax.imshow(missing_mask.T.values, aspect='auto', cmap='Greys', interpolation='nearest')
ax.set_yticks(range(len(opt_cols)))
ax.set_yticklabels([f"{s} {'PE' if c.endswith('PE') else 'CE'}" for s, c in zip(strikes, opt_cols)], fontsize=8)
ax.set_xlabel('Timestamp index (0 -> 974)')
ax.set_ylabel('Strike + Type')
ax.set_title('Missing data pattern: black = missing, white = present')
plt.tight_layout()
plt.show()
```

**What to take away from this plot:** The missing values are *scattered* — there's no obvious systematic pattern. No single column is mostly missing, no single timestamp is mostly missing. This is good news: it means for any missing cell, we likely have neighboring cells (same strike, nearby times) and same-time cells (nearby strikes) to learn from.

```python
## 4. Visualize the volatility surface as a heatmap
```

```python
fig, ax = plt.subplots(figsize=(13, 6))
iv_matrix = df[opt_cols].values   # rows=time, cols=strikes
iv_vis = np.clip(iv_matrix, 0, 0.5)   # clip extreme outliers just for visualisation
im = ax.imshow(iv_vis.T, aspect='auto', cmap='viridis', interpolation='nearest')
ax.set_yticks(range(len(opt_cols)))
ax.set_yticklabels([f"{s} {'PE' if c.endswith('PE') else 'CE'}" for s, c in zip(strikes, opt_cols)], fontsize=8)
ax.set_xlabel('Timestamp index')
ax.set_ylabel('Strike + Type')
ax.set_title('Implied volatility surface (yellow = high, dark = low, white = missing)')
plt.colorbar(im, ax=ax, label='Implied volatility')
plt.tight_layout()
plt.show()
```

**Huge insight in this plot:** Look at the right side of the heatmap (last ~10% of timestamps). The IV values explode (yellow). Everywhere else IV is calm (dark blue/teal).

**Why?** The expiry date is 27 Jan 2026, and the data ends on 27 Jan 2026. As options approach expiry, IV calculations become numerically unstable — small price changes translate into huge IV changes. This is a well-known phenomenon in options markets.

**Trap warning:** If we don't handle this near-expiry region carefully, our model will be confused. We'll need to either filter it out, model it separately, or use a transformation that compresses extreme values.

## 5. The volatility smile (or skew)

Pick a single timestamp and plot IV against strike. This is the most famous shape in options markets.

```python
# Pick the timestamp with the most data
best_row = df[opt_cols].notna().sum(axis=1).idxmax()
underlying = df.loc[best_row, 'underlying_price']
ivs = df.loc[best_row, opt_cols].values

fig, ax = plt.subplots(figsize=(11, 5))
colors = ['#D85A30' if c.endswith('PE') else '#185FA5' for c in opt_cols]
ax.scatter(strikes, ivs, s=70, c=colors, alpha=0.8)
ax.axvline(underlying, color='gray', linestyle='--', alpha=0.6)
ax.set_xlabel('Strike price')
ax.set_ylabel('Implied volatility')
ax.set_title(f'Volatility skew at {df.loc[best_row, "datetime"]}  |  spot = {underlying:.0f}')
ax.grid(True, alpha=0.3)

# Legend
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
ax.legend(handles=[
    Patch(facecolor='#D85A30', label='Put (PE)'),
    Patch(facecolor='#185FA5', label='Call (CE)'),
    Line2D([0], [0], color='gray', linestyle='--', label=f'Spot: {underlying:.0f}')
], loc='upper right')
plt.tight_layout()
plt.show()
```

**What this shape tells us:**
1. Puts (orange, left side) have *higher* IV than calls (blue, right side). This is the **volatility skew** — traders pay more for downside protection because crashes happen faster than rallies.
2. IV is *lowest* near the spot price (around 25500). This is the **at-the-money** point.
3. IV rises as you move away from spot in either direction (the 'smile' shape, though it's lopsided here).

**Why this matters for modeling:** At any timestamp, if we have most IVs and need to predict a missing one, we can use the smile shape. The missing IV at strike K should fit smoothly on the curve formed by the other strikes at the same time.

## 6. Time series view

How does IV for a single strike change over time? And how does it relate to the underlying price?

```python
fig, axes = plt.subplots(2, 1, figsize=(13, 7), sharex=True)

near_money_col = 'NIFTY27JAN2625900CE'
axes[0].plot(df['datetime'], df[near_money_col], color='#185FA5', linewidth=1, alpha=0.8)
axes[0].set_ylabel('Implied volatility')
axes[0].set_title(f'IV time series: {near_money_col}')
axes[0].grid(True, alpha=0.3)

axes[1].plot(df['datetime'], df['underlying_price'], color='#27500A', linewidth=1, alpha=0.8)
axes[1].set_ylabel('Underlying Nifty')
axes[1].set_xlabel('Datetime')
axes[1].set_title('Underlying price over the same window')
axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**Observations:**
- IV is mostly flat around 0.10–0.15 for most of the window, then spikes dramatically near expiry (Jan 27).
- The underlying Nifty drifts from ~26100 down to ~25000 — a sustained decline.
- The gaps in the plots are weekends and overnight hours (no trading).
- IV and underlying price are loosely *inversely* related — when Nifty fell, IV crept up. This is the classic 'leverage effect': market drops increase fear and option premiums.

## 7. Outliers and IV distribution

```python
all_iv = df[opt_cols].values.flatten()
all_iv = all_iv[~np.isnan(all_iv)]

print(f'Number of IV observations: {len(all_iv)}')
print(f'Min IV:   {all_iv.min():.4f}')
print(f'25th %:   {np.percentile(all_iv, 25):.4f}')
print(f'Median:   {np.median(all_iv):.4f}')
print(f'75th %:   {np.percentile(all_iv, 75):.4f}')
print(f'95th %:   {np.percentile(all_iv, 95):.4f}')
print(f'99th %:   {np.percentile(all_iv, 99):.4f}')
print(f'Max IV:   {all_iv.max():.4f}   <-- huge outlier!')
print()
print(f'Values above 1.0 (i.e. >100%): {(all_iv > 1.0).sum()} cells')
print(f'Values above 0.5 (i.e. >50%):  {(all_iv > 0.5).sum()} cells')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].hist(all_iv, bins=80, color='#534AB7', alpha=0.7)
axes[0].set_xlabel('Implied volatility (raw)')
axes[0].set_ylabel('Count')
axes[0].set_title(f'Raw IV distribution (max = {all_iv.max():.2f})')
axes[0].axvline(0.5, color='red', linestyle='--', label='0.5 threshold')
axes[0].legend()

axes[1].hist(all_iv[all_iv < 0.5], bins=80, color='#1D9E75', alpha=0.7)
axes[1].set_xlabel('Implied volatility (clipped at 0.5)')
axes[1].set_ylabel('Count')
axes[1].set_title('IV distribution after clipping outliers')
plt.tight_layout()
plt.show()
```

**Takeaway:** Most IV values are between 0.10 and 0.25 (10%–25%), which is normal for index options. A small number of cells have IV above 0.5 — these are almost all in the near-expiry region. We need to be careful with these when computing errors, because MSE is *very* sensitive to outliers (it squares the error).

## 8. Day 1 key insights (memorize these for the interview)

1. **Structure of the dataset:** 975 5-minute timestamps over 15 trading days, with 28 option columns (14 OTM puts below spot, 14 OTM calls above spot), all on Nifty50 with expiry 27 Jan 2026.

2. **What we predict:** ~20% of the cells are missing (5,460 of 27,300). The metric is MSE between our predictions and the hidden true values.

3. **The surface has clear structure we can exploit:**
   - *Cross-sectional smoothness*: at any time, IV varies smoothly with strike (the smile/skew shape).
   - *Temporal smoothness*: IV for a given strike changes slowly over consecutive 5-minute bars.
   - *Skew signature*: puts have higher IV than calls (downside fear premium).

4. **Missingness looks random** — there's no concentration of missing values in any column or time window. This means for any missing cell, we have nearby neighbors (in strike and time) to learn from.

5. **Biggest trap: near-expiry IV explosion.** The last ~10% of timestamps (close to expiry on 27 Jan) have IV values that spike to 1.0+ or even 5+. MSE is very sensitive to these outliers, so we must handle them deliberately (filter, clip, log-transform, or separate model).

6. **Lookahead bias rule:** When predicting an IV at timestamp t, we can use other strikes' IV *at the same time t* (the surface is cross-sectional), but we cannot use values from *future* timestamps. We can use values from past timestamps.
